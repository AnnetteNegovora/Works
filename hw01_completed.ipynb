#### **[task 6, 1б.]**
Познакомимся с рынком труда России. Для этого допишите функцию info_count.

```python
def info_count(df, col):
    ## группируем таблицу df по колонке col
    g = df.groupby(col)
    ## считаем количество значений в каждой категории колонки col через value_counts 
    ## делаем сортировку по количеству значений через sort_values
    ## normalize исключает значение  nan и считает долю, равносильно тому, что здесь хотят от функции
    g = df[col].value_counts(normalize = True).sort_values()
    
    ## далее строится график
    g.plot(kind='barh')
    for index, value in enumerate(g):
        plt.text(value, index, str(round(value*100,1))+'%')
    plt.grid(True)
    plt.show()
```

```python
plot_cols = ['town_type', 
            #'marital_status', 
            'professional_group',
            'finished_education', 
            #'age', 'region', 'wage'
            'sex', 
            'employed', 'industry',
           ]
for col in plot_cols:
    info_count(df, col)
```

Допишите функцию ниже для подсчета [медианной](https://ru.wikipedia.org/wiki/%D0%9C%D0%B5%D0%B4%D0%B8%D0%B0%D0%BD%D0%B0_(%D1%81%D1%82%D0%B0%D1%82%D0%B8%D1%81%D1%82%D0%B8%D0%BA%D0%B0)) [зарплаты](https://www.kommersant.ru/doc/4441454#:~:text=%D0%9C%D0%B5%D0%B4%D0%B8%D0%B0%D0%BD%D0%BD%D0%B0%D1%8F%20%D0%B7%D0%B0%D1%80%D0%BF%D0%BB%D0%B0%D1%82%D0%B0%20%E2%80%94%20%D1%8D%D1%82%D0%BE%20%D1%82%D0%B0%D0%BA%D0%BE%D0%B9%20%D1%83%D1%80%D0%BE%D0%B2%D0%B5%D0%BD%D1%8C,%D1%81%D0%BD%D0%B8%D0%B6%D0%B5%D0%BD%D0%B8%D0%B5%20%D1%81%D1%80%D0%B5%D0%B4%D0%BD%D0%B8%D1%85%20%D0%B7%D0%B0%D1%80%D0%BF%D0%BB%D0%B0%D1%82%20%D0%B2%20%D1%81%D1%82%D1%80%D0%B0%D0%BD%D0%B5.) и построения графика.

```python
def info_wage_median(df, col):
    ## группируем таблицу df_emp по колонке col
    ## считаем медианное значение в колонке 'wage'
    ## делаем сортировку по значениям
    ## делим все значения на тысячу, чтобы смотреть на зарплату в тысячах
    ## результат записываем в переменную g
    d = (df.groupby([col])
        .agg(medium_wage = ('wage', np.median))
           )
    g = d['medium_wage']/1000
    g = g.astype(float)
    g = g.sort_values()
    #new_col = df.groupby(col)['wage'].median()
    #new_col = new_col.sort_values()
    #g = new_col / 1000

    #c = (df.groupby(col)
    #    .agg(median_wage = ('wage', df['wage'].median())
     #     )
    #g = c /1000
    #ad = df.groupby(['sex'])
    #ad = df['wage'].median()

    ## далее строится график
    fig, ax = plt.subplots(figsize = (12,6))
    g.plot(kind='barh')
    for index, value in enumerate(g):
        plt.text(value, index, str(int(value)))
    plt.grid(True)
    plt.xlabel('тыс. руб.')
    plt.show()

for col in plot_cols:
    info_wage_median(df_emp, col)

```

### Распределение зарплат

#### **[task 11, 1.5б.]**

Добавить в переменную `df_emp` новый столбец `wage_bin`, где в соответствии с зарплатой будет указан диапазон зарплаты в тыс. рублей. Возможные значения диапазон начинаются от 0 и идут с шагом 4. Например, для зарплаты 2000 рублей должен быть указан диапазон (0,4], а для 5000 рублей – (4,10], и т.д.

```python
df_emp['wage/1000'] = df_emp['wage']/1000
# создаем новый столбец, где указывааем зп в тыс.руб.
n = int(max(df_emp['wage/1000'])//4)
# ищем количество интервалов, целочисленно поделя максимальную указанную зп в тыс.руб. на 4(шаг)
bins = [0]*(n+1)
# создаем пустой список из n+1 элементов и наполняем его
for i in range(n+1):
        bins[i]=4*i
df_emp['wage_bin'] = pd.cut(df_emp['wage/1000'], bins)
df_emp.head()
```

#### **[task 12, 1б.]**
Посчитайте долю каждого диапазона зарплат в выборке, запишите результат в виде Pandas.Series в переменную g.

```python
g = df_emp['wage_bin'].value_counts(sort=False)
#убираем сортировку, чтобы нагляднее было видно распределение по интервалам
sum_znach = g.sum()
g = g/sum_znach # находим долю
g = g[(g!=0)]
g
#убираем все нулевые значения 
```
